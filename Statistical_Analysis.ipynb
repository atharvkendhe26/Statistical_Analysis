# Hypothesis 

#### 1. AutoDrive Components Ltd. – Engine Assembly Time Evaluation

```python
# State Hypothesis
#H0
#H1
```

```python
theory_mean = 1.8+0.02*450
mean = 10.8
std = 25
sample = 35
alpha = 0.05
```

```python
#tstats
standard_error = std/np.sqrt(sample)
t_statistic =(mean-theory_mean)/standard_error
t_statistic
```

```python
from scipy.stats import norm
```

```python
# Critical Value
critical_value = norm.ppf(1-alpha)
critical_value
```

```python
if t_statistic > critical_value:
    decision = "H1"
else:
    decision ="H0"
```

```python
decision
```

```python
# From the above AutoDrive Componantents example we can see that there is no change 
```

#### 2. FreshBake Foods Pvt. Ltd. – Baking Time Evaluation

```python
# Stats Hypothesis
#H0
#H1
```

```python
theory_mean1 = 10 +0.08*75
mean1 = 16.2
std1 = 6
sample1 = 30
alpha1 = 0.05
```

```python
#tstats
standard_error1 = std/np.sqrt(sample1)
t_statistic1 = (mean1-theory_mean1)/standard_error1
t_statistic1
```

```python
# Critical Value
critical_value1 = norm.ppf(1-alpha1)
critical_value1
```

```python
if t_statistic1 > critical_value1 :
    decision="H1"
else:
    decision="H0"
```
